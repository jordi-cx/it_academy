### M7 T13
La nostra empresa es dedica a realitzar anàlisis de dades hospitalaries, i un client ens ha sol·licitat una eina per identificar dades bàsiques dels hospitals de cada comunitat autònoma de forma automàtica.

Per això, cal crear una funció que rebi dos paràmetres:

- Un DataFrame amb informació d’hospitals (com el del fitxer hospitals.csv).
- El nom d’una comunitat autònoma (com a cadena de text).

La funció ha de retornar un diccionari amb les següents dades relacionades amb la comunitat autónoma indicada:

- Quantitat d’hospitals que hi ha a la comunitat autònoma
- Quantitat d’hospitals que hi ha a la comunitat autónoma que tenen més del 50% d’ocupació
- Indice_satifaccion màxim de la comunitat autónoma
- Indice_satifaccion mínim de la comunitat autònoma
- Total de llits a la comunitat autónoma
- Mitjana de l’index de satisfacció dels hospitals de la comunitat autònoma

Finalment, carrega el fitxer adjunt amb aquesta instrucció: data = pd.read_csv("arxiu csv", sep = ";", encoding = "latin-1") 
i executa la funció per a diverses comunitats autònomes i mostra els resultats obtinguts.

```python
data = pd.read_csv("hospitals.csv", sep = ";", encoding = "latin-1")
data.head()
```

```python
data.tail()
```

```python
def dades_hospitals(dfh, comunitat):
    dicc_dades = {}

    # Quantitat d'hospitals a la comunitat autònoma
    num_hospitals = dfh.loc[dfh['comunidad_autonoma'] == comunitat, 'hospital_id'].count()
    dicc_dades['num_hospitals'] = int(num_hospitals)

    # Quantitat d'hospitals a la comunitat autònoma amb més del 50% d'ocupació
    num_h_ocupats = dfh.loc[(dfh['comunidad_autonoma'] == comunitat) & (dfh['porcentaje_ocupacion'] > 50.0), 'hospital_id'].count()
    dicc_dades['num_h_ocupats'] = int(num_h_ocupats)

    # Índex satisfacció màxim de la comunitat autònoma
    max_satisfaccio = dfh.loc[dfh['comunidad_autonoma'] == comunitat, 'indice_satisfaccion'].max().round(2)
    dicc_dades['max_satisfaccio'] = float(max_satisfaccio)

    # Índex satisfacció mínim de la comunitat autònoma
    min_satisfaccio = dfh.loc[dfh['comunidad_autonoma'] == comunitat, 'indice_satisfaccion'].min().round(2)
    dicc_dades['min_satisfaccio'] = float(min_satisfaccio)

    # Total de llits a la comunitat autònoma
    total_llits = dfh.loc[dfh['comunidad_autonoma'] == comunitat, 'num_camas'].sum()
    dicc_dades['total_llits'] = int(total_llits)

    # Mitjana de l'índex de satisfacció
    mitjana_satisfaccio = dfh.loc[dfh['comunidad_autonoma'] == comunitat, 'indice_satisfaccion'].mean().round(2)
    dicc_dades['mitjana_satisfaccio'] = float(mitjana_satisfaccio)

    return dicc_dades

print(dades_hospitals(data, 'Andalucia'))
print(dades_hospitals(data, 'Cataluña'))
print(dades_hospitals(data, 'Comunidad de Madrid'))
```

### M7 T14
La nostra empresa está colaborant amb el colectiu de metges per veure quina càrrega de treball están patint. La idea es veure quins metges han de donar suport a més pacients i quins sembla que tenen menys persones a càrrec.

Per fer-ho, ens han proporcionat un conjunt de dades (hospitals.csv) que conté informació detallada sobre els hospitals del país.

Crea una funció en Python que rebi com a paràmetres:

- Un DataFrame amb dades d’hospitals.
- El nombre n d’hospitals que es volen obtenir com a resultat.

La funció ha de calcular, per a cada hospital la càrrega de treball dels metges, utilitzant la següent fórmula:

càrrega treball metge = num_llits /num_metges * percentatge ocupació

La funció ha de retornar els n hospitals amb més càrrega de treball per metge.

Executa la funció per obtenir 10 hospitals i després per obtenir-ne 20.

```python
def carrega_hospitals(dfh, n):
    dfh['carga_medico'] = ((dfh['num_camas'] / dfh['num_medicos']) * dfh['porcentaje_ocupacion'] / 100).round(2)
    dfh = dfh.sort_values(by='carga_medico', ascending=False)
    # return dfh.head(n)
    return dfh.iloc[0:n, [1, 2, 10]]

carrega_hospitals(data, 10)
```

```python
carrega_hospitals(data, 20)
```

## PER PRACTICAR MÉS

```python
hospitals = pd.read_csv('hospitals.csv', sep = ";", encoding = "latin-1")
hospitals
```

### FILTRES:

```python
# a) Mostra totes les dades dels hospitals que tenen entre 100 i 120 "num_camas".
hospitals[(hospitals['num_camas'] >= 100) & (hospitals['num_camas'] <= 120)]
```

```python
# b) Mostra els noms dels hospitals que estan a Barcelona - Catalunya.
hospitals[(hospitals['localidad'] == 'Barcelona') & (hospitals['comunidad_autonoma'] == 'Cataluña')]
```

```python
# c) Mostra els 3 hospitals amb més "indice_satisfaccion" del grup d'hospitals de Madrid i Andalusia.
hospitals[hospitals['comunidad_autonoma'].isin(['Comunidad de Madrid', 'Andalucia'])].sort_values(by='indice_satisfaccion', ascending=False).head(3)
```

```python
# d) Mostra els hospitals que comencen per lletra "C" i terminen per lletra "a".
hospitals[hospitals['nombre'].str.startswith('C') & hospitals['nombre'].str.endswith('a')]
```

### AGRUPACIONS:

```python
# a) Mostra el "indice_satisfaccion" mig de cada comunitat autònoma i província de la taula "hospitals".
hospitals.groupby(['comunidad_autonoma','provincia'])['indice_satisfaccion'].mean().round(2)
```

```python
# b) Mostra el "num_camas" total de cada comunitat autònoma.
hospitals.groupby('comunidad_autonoma')['num_camas'].sum().reset_index(name = 'Num_camas')
```

```python
# c) Mostra el "porcentaje_ocupacion" més petit de cada província de cada comunitat autònoma.
hospitals.groupby(['comunidad_autonoma','provincia'])['porcentaje_ocupacion'].min()
```

```python
# d) Mostra quantes províncies i localitats té cada comunitat autònoma.
hospitals_prov = hospitals.groupby('comunidad_autonoma').agg({'provincia': 'nunique', 'localidad': 'nunique'})
hospitals_prov
```

```python
# e) Mostra les comunitats autònomes que tenen menys de 5 hospitals.
hospitals_cc = hospitals.groupby('comunidad_autonoma')['hospital_id'].count().reset_index(name = 'Num_hospitals')
hospitals_cc[hospitals_cc['Num_hospitals'] < 5]
```
