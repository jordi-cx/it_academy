### M7 T8
Crea una funció que:

1. Demani a l’usuari 3 noms de ciutats, un per un.
2. Comprovi si cada ciutat conté la lletra 'E' o 'e' i té més de 5 lletres.
3. Si la ciutat compleix aquestes condicions, afegeix-la a una llista.
4. Quan hagi acabat, retorna la llista completa de ciutats que compleixen els requisits.

Finalment, crida la funció i imprimeix la llista completa.

```python
def tria_ciutats():
  ciutats = []

  for i in range(3):
    ciutat = input("Digues alguna ciutat: ")
    
    if 'e' in ciutat.lower() and len(ciutat) > 5:
      ciutats.append(ciutat)
  
  return ciutats

print(tria_ciutats())
```

### M7 T9
Crea una funció que rebi per paràmetre el nombre de colors que l’usuari vol introduir. A continuació:

1. Demana els colors un per un amb un missatge del tipus: "Introdueix el color 1", "Introdueix el color 2", etc.

3. Quan tots els colors hagin estat introduïts, mostra la llista completa.

Finalment, abans de cridar la funció, es demana a l’usuari quants colors vol introduir.

```python
def mostra_colors(num_colors):
    colors = []

    for i in range(num_colors):
        # colors.append(input(f"Introduce el color {i+1}: "))
        prompt = "Digues el color " + str(i + 1) + ": "
        color = input(prompt)
        colors.append(color)

    print(colors)

num = int(input("Quants colors vols introduir? "))
mostra_colors(num)

```

### M7 T10

### M7 T11
Escriu un programa que faci el següent:

1. Demani a l'usuari que introdueixi una frase.

2. Utilitza una funció que rebi aquesta frase com a argument i retorni un diccionari amb:
- Les paraules que apareixen a la frase com a claus.
- El nombre de vegades que apareixen com a valors.

No utilitzis la funció integrada count() per comptar les paraules.

3. Finalment, mostra el diccionari per pantalla.

```python
# Funció auxiliar per netejar la frase de signes de puntuació, mantenint espais
def neteja_frase(text):
    frase_neta = ""

    for char in text:
        if char.isalpha() or char.isnumeric() or char == " ":
            frase_neta = frase_neta + char.lower()

    print(frase_neta)
    return(frase_neta)
```

```python
frase_usuari = input("Escriu una frase: ")

def paraules_diferents(frase):
    frase = neteja_frase(frase)

    # Llista de paraules
    paraules = frase.split()
    print(paraules)

    # Diccionari amb comptador de paraules
    dicc_comptador = {}

    for paraula in paraules:
      if paraula not in dicc_comptador:
        dicc_comptador[paraula] = 1
      else:
        dicc_comptador[paraula] += 1

    return dicc_comptador

print(paraules_diferents(frase_usuari))
```

```python
def compta_vocals(frase):
    vocals = 0
    
    for lletra in frase:
        if lletra in ['a', 'e', 'i', 'o', 'u']:
            vocals += 1
     
    return vocals

text = input("Digues una frase o paraula: ")
num_vocals = compta_vocals(text)

print("El teu text conté", num_vocals, "vocals")
```

3. Defineix una funció que, donada una frase, retorni el nombre de paraules úniques que conté, sense fer servir la funció len(). (Pots utilitzar el mètode split() d’un string.)

```python
def compta_paraules(text):
    llista_paraules = text.split()
    llista_uniques = []
    num_paraules = 0
    
    for paraula in llista_paraules:
        if paraula not in llista_uniques:
            llista_uniques.append(paraula)
            num_paraules += 1
    
    print(llista_uniques)
    return num_paraules

frase = input("Escriu una frase: ")
num_uniques = compta_paraules(frase)
print("El teu text conté", num_uniques, "paraules diferents.")
```

5. Escriu una funció que demani un text i retorni un diccionari amb les lletres que apareixen i la quantitat de cops que apareixen, sense fer servir funcions d’agregació com count().

```python
def compta_lletres(text):
    dicc_lletres = {}

    # Inicialitza diccionari de lletres
    for lletra in text:
        if lletra.isalpha():
            dicc_lletres[lletra] = 0
    
    # Compta cada lletra
    for lletra in text:
        if lletra.isalpha():
            dicc_lletres[lletra] += 1
    
    return dicc_lletres

frase = input("Escriu una frase o paraula: ")
print(compta_lletres(frase))
```

```python
def plats_calories():
    dicc_calories = {}

    for i in range(5):
        plat = input("Digues un plat: " )
        calories = int(input("Quantes calories té això? "))
        dicc_calories[plat] = calories
    
    return dicc_calories

diccionari_calories = plats_calories()
print(diccionari_calories)
```
