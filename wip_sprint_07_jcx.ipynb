## N1.1: Selecció de Personal

**ALGORITMO PRINCIPAL**: *main_seleccio()*

Mientras no llegamos a 10 candidaturas:
1. Obtener datos de canditatura: *obtenir_dades()*
2. Validar los datos de la candidatura: *validar_dades()*
3. Calcular la puntuación de la candidatura: *calcular_puntuacio()*
4. Mostrar el informe individual de la candidatura: *mostrar_individual()*

Al llegar a 10 candidaturas validadas:

5. Ordenar candidaturas para informe final: *crear_informe()*
6. Mostrar informe final: *mostrar_informe()*


```python
def obtenir_dades(dades_test=None):
  """"
  Esta función tiene como entrada una tupla de datos de un posible candidato.
  Si nos proporcionan datos de entrada, utilizaré esos datos.
  Si no hay datos de entrada, la función pedirá los datos por pantalla al usuario.

  Queremos incorporar los siguientes datos:
  - Nombre y Apellidos (nom)
  - Teléfono de contacto (telefono)
  - Años de experiencia (experiencia)
  - Conocimientos de Catalán (catala)
  - Dominio de Excel (excel)
  - Título Académico (grau)

  La función retornará una tupla de datos con seis campos:
  (nom, telefon, experiencia, catala, excel, grau)
  """

  if dades_test:
    print("\nCarregant les dades automàticament...")
    try:
      nom = dades_test[0]
      telefon = dades_test[1]
      experiencia = dades_test[2]
      catala = dades_test[3]
      excel = dades_test[4]
      grau = dades_test[5]
      return (nom, telefon, experiencia, catala, excel, grau)

    except IndexError:
      print("Error: Falten dades")
      return
    except:
      print("Error: Dades incorrectes")
      return

  else:
    print("\nSisplau, entra les dades manualment...")
    nom = input("Introdueix el nom: ").strip()
    telefon = input("Introdueix el telèfon: ").strip()
    experiencia = int(input("Introdueix el número d'anys a l'empresa: ").strip())
    catala = int(input("Introdueix el teu nivell de català\n(0 = sense coneixement, 1 = fins a nivell B1, 2 = nivell B2, 3 = C1 o superior): ").strip())
    excel = int(input("Introdueix el teu nivell d'Excel\n(0 = nivell bàsic, 1 = ús de funcions, 2 = taules dinàmiques, 3 = Visual Basic i Macros): ").strip())
    grau = int(input("Introdueix el teu nivell de Títol acadèmic\n(0 = ESO, 1 = FP, 2 = Universitat, 3 = Màster o superior): ").strip())
    return (nom, telefon, experiencia, catala, excel, grau)
```

```python
def validar_dades(dades_candidat):
  """
  Esta función recibe una tupla de datos
  y comprueba que estos datos sean del tipo correcto,
  para cada uno de sus campos:

  nom -> Debe ser un texto no vacío
  telefon -> Debe ser un texto no vacío
  experiencia -> Debe ser un número entero
  catala -> Debe ser un número entero entre 0 y 3
  excel -> Debe ser un número entero entre 0 y 3
  grau -> Debe ser un número entero entre 0 y 3

  Finalmente la función retorna True si los datos están correctos,
  o en caso de datos incorrectos, retorna False.
  """

  try:
    nom = dades_candidat[0]
    telefon = dades_candidat[1]
    experiencia = dades_candidat[2]
    catala = dades_candidat[3]
    excel = dades_candidat[4]
    grau = dades_candidat[5]
  except:
    print("Error: Dades incorrectes")
    return False

  if len(nom) == 0:
    print("Error: Nom no vàlid")
    return False
  if len(telefon) == 0:
    print("Error: Telèfon no vàlid")
    return False

  if (catala < 0 or catala > 3) or (excel < 0 or excel > 3) or (grau < 0 or grau > 3):
    print("Error: Valors de Català / Excel / Títol Acadèmic han d'estar entre 0 i 3")
    return False

  return True
```

```python
def calcular_puntuacio(experiencia, catala, excel, grau):
  """
  Esta función recibe cuatro argumentos de entrada de tipo entero:
  - Años de experiencia: experiencia
  - Nivel de catalán: catala
  - Dominio de Excel: excel
  - Título académico: grau

  La función retorna un número según la fórmula:
  puntuación = 1.5 * experiencia + catala + excel + grau
  """

  if (experiencia < 0):
    print("Error: El número d'anys d'experiència ha de ser major que 0")
    return

  if (catala < 0 or catala > 3) or (excel < 0 or excel > 3) or (grau < 0 or grau > 3):
    print("Error: Valors de Català / Excel / Títol Acadèmic han d'estar entre 0 i 3")
    return

  return 1.5 * experiencia + catala + excel + grau
```

```python
def classificar_candidat(puntuacio):
  """
  Con el valor de entrada proporcionado del parámetro *puntuacio* (tipo float),
  la función retorna un texto según la clasificación del candidato:

  'Prioritat Alta' si puntuacio > 12
  'Candidat Adequat' si 8 ≤ puntuacio < 12
  'No Prioritari' si puntuacio < 8
  """

  if puntuacio > 12:
    return "Alta Prioritat"
  elif 8 <= puntuacio < 12:
    return "Candidat Adequat"
  else:
    return "No Prioritari"
```

```python
def mostrar_individual(dades_canditat):
  """
  Esta función recibe como entrada un diccionario
  con los datos ya validados correctamente de un posible candidato.
  El diccionario incluye las siguientes claves y valores:

  - 'nom' -> text con el nombre y apellidos
  - 'telefon' -> teléfono como string
  - 'experiencia' -> número entero con los años de experiencia
  - 'catala' -> número entero con el nivel de catalán
  - 'excel' -> número entero con el nivel de Excel
  - 'grau' -> número entero con el nivel de grado académico
  - 'puntuacio' -> número float con la puntuación del candidato
  - 'classificacio' -> string con la valoración del candidato

  Se muestra por pantalla un texto con los datos del candidato.
  """

  informe = ""

  try:
    for key, value in dades_canditat.items():
      if key == 'nom':
        informe += f"Nom i Cognoms: {value}\n"
      elif key == 'telefon':
        informe += f"Telèfon: {value}\n"
      elif key == 'experiencia':
        informe += f"Anys d'experiència a l'empresa: {value}\n"

      elif key == 'catala':
        nivell_catala = ''
        if value == 0:
          nivell_catala = 'Sense coneixement'
        elif value == 1:
          nivell_catala = 'Nivell B1'
        elif value == 2:
          nivell_catala = 'Nivell B2'
        elif value == 3:
          nivell_catala = 'C1 o superior'
        informe += f"Coneixement de català: {nivell_catala}\n"

      elif key == 'excel':
        nivell_excel = ''
        if value == 0:
          nivell_excel = 'Nivell bàsic'
        elif value == 1:
          nivell_excel = 'Ús de funcions'
        elif value == 2:
          nivell_excel = 'Taules dinàmiques'
        elif value == 3:
          nivell_excel = 'Visual Basic i Macros'
        informe += f"Domini d'Excel: {nivell_excel}\n"

      elif key == 'grau':
        nivell_academic = ''
        if value == 0:
          nivell_academic = 'ESO'
        elif value == 1:
          nivell_academic = 'FP'
        elif value == 2:
          nivell_academic = 'Universitat'
        elif value == 3:
          nivell_academic = 'Màster o superior'
        informe += f"Títol Acadèmic: {nivell_academic}\n"

      elif key == 'puntuacio':
        informe += f"Puntuació: {value}\n"
      elif key == 'classificacio':
        informe += f"Classificació: {value}"

    print(informe)

  except:
    print("Error creant infore: dades incorrectes")
```

```python
def mostrar_informe(candidatures):
  """
  Esta función recibe una lista ordenada con las 10 candidaturas seleccionadas.
  Es una lista de diccionarios que incluyen los datos de cada candidatura
  - 'nom' -> text con el nombre y apellidos
  - 'telefon' -> teléfono como string
  - 'puntuacio' -> número float con la puntuación

  Se muestra un texto con el informe formateado de los datos de los candidatos.
  """
  informe = '\nINFORME FINAL\n'
  for i in range(len(candidatures)):
    informe += f"{i+1}) Nom: {candidatures[i]['nom']}  Telèfon: {candidatures[i]['telefon']}  Puntuació: {candidatures[i]['puntuacio']}\n"

  print(informe)
```

**ALGORITMO PRINCIPAL**: *main_seleccio()*

Mientras no llegamos a 10 candidaturas:
1. Obtener datos de canditatura: *obtenir_dades()*
2. Validar los datos de la candidatura: *validar_dades()*
3. Calcular la puntuación de la candidatura: *calcular_puntuacio()*
4. Calcular clasificación de candidatura: *classificar_candidat()*
5. Añadir candidato a la lista de candidaturas
6. Mostrar el informe individual de la candidatura: *mostrar_individual()*

Al llegar a 10 candidaturas validadas:

7. Ordenar candidaturas para informe final: *crear_informe()*
8. Mostrar informe final: *mostrar_informe()*


```python
def main_seleccio():

  inputs_test = [
    "Laura Serra", "608112233", 6, 3, 3, 3,
    "Clara Gómez", "655221199", 3, 3, 3, 2,
    "Sergio López", "644112211", 3, 2, 2, 2,
    "Candidat Error", "611000000", 5, 7, 3, 2, # Català = 7 → NO VÀLID
    "Marta Ruiz", "699887766", 2, 3, 2, 1,
    "Marc Vidal", "612443322", 6, 3, 3, 3,
    "Júlia Romero", "677998877", 5, 3, 3, 2,
    "Pol Navarro", "699554433", 4, 3, 2, 2,
    "David Ferrer", "677665544", 2, 1, 1, 1,
    "Anna Paredes", "633221144", 1, 2, 1, 0,
    "Anibal Paredes", "644272727", 1, 2, 1, 0,
    "Núria Casals", "600112200", 8, 2, 3, 3 # 11è candidat
    ]

  llista_candidats = []

  while len(llista_candidats) < 10:
    # 1
    candidat = obtenir_dades(inputs_test[:6])
    del inputs_test[:6]

    # 2
    if validar_dades(candidat):
      dicc_candidat = {}
      dicc_candidat['nom'] = candidat[0]
      dicc_candidat['telefon'] = candidat[1]
      dicc_candidat['experiencia'] = candidat[2]
      dicc_candidat['catala'] = candidat[3]
      dicc_candidat['excel'] = candidat[4]
      dicc_candidat['grau'] = candidat[5]

      # 3
      dicc_candidat['puntuacio'] = calcular_puntuacio(candidat[2], candidat[3], candidat[4], candidat[5])
      # 4
      dicc_candidat['classificacio'] = classificar_candidat(dicc_candidat['puntuacio'])
      # 5
      llista_candidats.append(dicc_candidat)
      # 6
      print("\nINFORME CANDIDATURA", len(llista_candidats))
      mostrar_individual(dicc_candidat)

    if len(llista_candidats) < 10:
      input("\nPrem ENTER per entrar següent candidat...")
    else:
      input("\nPrem ENTER per veure informe final...")

  # 7
  llista_candidats.sort(key=lambda x: (x['classificacio'], -x['puntuacio']))
  # 8
  mostrar_informe(llista_candidats)
```

```python
def obtenir_segment_i_factor(client):
  """
  Esta función recibe como entrada un diccionario con los datos iniciales de un cliente.

  Incorporaremos al cliente los campos (clave / valor) de:
  - 'segment': Un string con la combinación del tipo de comportamiento con el grupo de edad (ej. Navega pero no añade productos / 18-25)
  - 'factor_codi': Número float correspondiente según su código de comportamiento (lo obtenemos del diccionario_marketing)

  La función retorna un diccionario con los datos actualizados del cliente.
  """

  cliente_actualizado = client.copy()
  cliente_actualizado['segment'] = dicc_marketing['comportaments'][client['codi_comportament']] + " / " + client['grup_edat']
  cliente_actualizado['factor_codi'] = dicc_marketing['factors_comportament'][client['codi_comportament']]

  return cliente_actualizado
```

```python
def calcular_risc(factor_comportament, factor_grup_edat):
  """
  Esta función recibe como parámetros de entrada dos números float (factor_comportament, factor_grup_edat)
  para poder calcular y retornar el valor de riesgo que retornaremos según la fórmula:
  riesgo = factor_comportament * factor_grup_edat
  """
  return round(factor_comportament * factor_grup_edat, 2)
```

```python
def redactar_missatge(risc, nom):
  """
  Esta función recibe un float con el factor de riesgo y el nombre del cliente.
  Retorna un string con el mensaje de texto adecuado para su segmento.

  Haciendo una primera inspección de los valores posibles de riesgo y segmento,
  propongo estos posibles casos distintos:

  0: No añaden productos => (riesgo <= 0.15) => missatge_0
  1: No llegan al checkout => (0.15 < riesgo <= 0.30) => missatge_1
  2-3: No llegan a pagar => (0.30 < riesgo <= 0.60) => missatge_2
  3-4: Cesta de la compra abandonada => (0.60 < riesgo <= 0.80) => missatge_3
  3-4: Cesta de la compra abandonada (jóvenes) => (riesgo > 0.80) => missatge_4
  """

  missatge_0 = "Gracias por su visita:\nEsperemos encuentre todo lo que necesita en nuestra tienda."
  missatge_1 = "Gracias por su visita:\nRecuerde hacer checkout para procesar su compra."
  missatge_2 = "Gracias por comprar en nuestra tienda:\nRecuerde completar el checkout para poder procesar su pedido."
  missatge_3 = "Gracias por comprar en nuestra tienda:\nPor favor recuerde hacer checkout para poder completar su compra."
  missatge_4 = f"Buenos días {nom}:\nEstamos muy contentos de contar contigo como cliente.\nIMPORTANTE: por favor, recuerda hacer el checkout para poder completar tu compra.\nMuchas gracias, deseamos que encuentres todo lo que deseas en nuestra tienda!"

  match risc:
    case x if x <= 0.15:
      return missatge_0
    case x if 0.15 < x <= 0.30:
      return missatge_1
    case x if 0.30 < x <= 0.60:
      return missatge_2
    case x if 0.60 < x <= 0.80:
      return missatge_3
    case x if x > 0.80:
      return missatge_4
    case _:
      return missatge_0
```

```python
def processar_clients(llista_clients):
  """
  Esta función recibe como entrada una lista de diccionarios con los datos iniciales de cada cliente.

  Cada diccionario viene con los siguientes campos de clave / valor:
    - 'id_client': Número entero con el identificador único de cada cliente
    - 'nom': String con el nombre
    - 'edat': Número entero con la edad
    - 'grup_edat': String con el grupo de edad
    - 'codi_comportament': Número entero con el código de comportamiento según los valores:
        0 -> Navega pero no añade productos
        1 -> Añade productos pero no llega al checkout
        2 -> Llega al checkout pero no paga
        3 -> Cesta de la compra abandonada hace menos de 24h
        4 -> Cesta de la compra abandonada hace más de 24h

  Al procesar cada cliente, debemos incorporar los siguientes campos:
    - 'segment': Un string con la combinación del tipo de comportamiento con el grupo de edad (ej. Navega pero no añade productos / 18-25)
    - 'factor_codi': Número float correspondiente según su código de comportamiento (lo obtenemos del diccionario_marketing)
    - 'factor_edat': Número float correspondiente según su grupo de edad (lo obtenemos del diccionario_marketing)
    - 'risc_final': Número float con el cálculo del riesgo según la fórmula risc_final = factor codi * factor edat
    - 'missatge': Un string con el mensaje de texto adecuado para su segmento

  Se retorna una lista con los datos actualizados de los clientes.
  """
  clients_mktg = []

  for client in llista_clients:
    client_plus = client.copy()
    client_plus = obtenir_segment_i_factor(client)
    client_plus['factor_edat'] = obtenir_factor_edat(client)
    client_plus['risc_final'] = calcular_risc(client_plus['factor_codi'], client_plus['factor_edat'])
    client_plus['missatge'] = redactar_missatge(client_plus['risc_final'], client_plus['nom'])
    clients_mktg.append(client_plus)

  return clients_mktg
```

```python
def mostrar_diccionari(client):
  """
  Esta función recibe como entrada un diccionario con todos los datos actualizados de un cliente
  y lo imprime por pantalla.
  """

  for key, value in client.items():
    custom_sep = ": " if not key == 'missatge' else ":\n"
    print(key, value, sep = custom_sep)

  print()
```

## N2.1: Gestió de Risc i Priorització de Pacients

ALGORITME PRINCIPAL: *main_gestio_pacients()*
1. Obtener datos de pacientes: *obtenir_dades_pacients()*

Para cada paciente:

2. Convertir datos de paciente: *convertir_dades()*
3. Limpiar datos de síntomas: *normalitzar_simptomes()*
4. Asignar valores de riesgo a cada síntoma: *assignar_risc_simptomes()*
5. Calcular valor de riesgo total: *calcular_risc_total()*
6. Mostrar datos correctos de paciente: *mostrar_pacient()*

Finalmente:

7. Ordenar la lista de pacientes por prioridad: *ordenar_pacients()*
8. Mostrar lista ordenada de pacientes: *mostrar_llista_pacients()*


```python
def obtenir_dades_pacients():
  pacients = [
      "Maria|45|tos, Fiebre Alta, dolor_pecho",
      "Luis|33|dolor_cabeza,faTiga",
      "Sara|67|fiebre alta, dificultad RESPIRAR, tos",
      "Jordi|52|fatiga , tos",
      "Anna|29|fiebre alta,dolor_cabeza"
  ]

  return pacients
```

```python
obtenir_dades_pacients()
```

```python
def convertir_dades(dades):
  """
  Esta función tiene como entrada un string con los datos en bruto de un paciente (separados mediante |).
  Retorna un diccionario con los datos del paciente procesados.
  """
  try:
    nom, edat, simptomes = dades.split("|")

    return {
        'nom': nom,
        'edat': edat,
        'simptomes': simptomes
    }

  except ValueError:
    print("Error: Datos incorrectos del paciente.")
    return None

  except AttributeError:
    print("Error: Datos de entrada incorrectos.")
    return None
```

```python
convertir_dades('Maria|45|tos, Fiebre Alta, dolor_pecho')
```

```python
def normalitzar_simptomes(simptomes):
  simptomes = simptomes.strip().lower()

  simptomes_ok = []
  for simptoma in simptomes.split(","):
    simptoma = simptoma.strip().replace(" ", "_")
    simptomes_ok.append(simptoma)

  return simptomes_ok
```

```python
normalitzar_simptomes('tos, Fiebre Alta, dolor_pecho')
```

```python
def assignar_risc_simptomes(simptomes):
  """
  Esta función recibe como entrada una lista de simptomes ya normalizada.

  Cada elemento de la lista de simptomes es un string con el nombre correcto.

  Convertiremos cada elemento de la lista en una tupla con dos elementos:
  (simptoma, valor de riesgo según el diccionario nivells_risc)

  Retornamos la lista de tuplas.
  """

  nivells_risc = {
      "fiebre_alta": 3,
      "dificultad_respirar": 5,
      "dolor_pecho": 4,
      "tos": 1,
      "fatiga": 1,
      "dolor_cabeza": 1
  }

  simptomes_risc = []

  for simptoma in simptomes:
    if simptoma in nivells_risc:
      simptomes_risc.append((simptoma, nivells_risc[simptoma]))
    else:
      simptomes_risc.append((simptoma, 0))

  return simptomes_risc
```

```python
assignar_risc_simptomes(['tos', 'fiebre_alta', 'dolor_pecho'])
```

```python
def calcular_risc_total(simptomes):
  """
  Esta función recibe una lista de tuplas de simptomas y sus respectivos valores de riesgo.
  Retorna el valor de riesgo total.
  """
  risc_total = 0

  for simptoma_risc in simptomes:
    if simptoma_risc[1] > 0:
      risc_total += simptoma_risc[1]

  return risc_total
```

```python
calcular_risc_total([('tos', 1), ('fiebre_alta', 3), ('dolor_pecho', 4)])
```

```python
def mostrar_pacient(pacient):
  """
  Esta función recibe como entrada un diccionario con los datos de un paciente.
  Muestra por pantalla los datos del paciente.
  """
  print(f"Nombre: {pacient['nom']}")
  print(f"Edad: {pacient['edat']}")
  print(f"Síntomas: {pacient['simptomes']}")
  print(f"Riesgo total: {pacient['risc_total']}")
  print()
```

```python
def ordenar_pacients(pacients):
  """
  Esta función recibe como entrada una lista de diccionarios con los datos de los pacientes.
  Retorna la lista de pacientes ordenada por su riesgo de mayor a menor.
  """
  pacients.sort(key=lambda x: x['risc_total'], reverse=True)
  return pacients
```

```python
def mostrar_llista_pacients(pacients):
  """
  Esta función recibe como entrada una lista de diccionarios con los datos de los pacientes.
  Muestra por pantalla la lista de pacientes formateada para que sea accesible visualmente.
  """
  print("LISTA DE PACIENTES SEGÚN PRIORIDAD:")
  for pacient in pacients:
    print(f"Nombre: {pacient['nom']}\tEdad: {pacient['edat']}\tRiesgo: {pacient['risc_total']}")
```

```python
def main_gestio_pacients():
  dades_pacients = obtenir_dades_pacients()
  pacients = []

  for dades in dades_pacients:
    pacient = convertir_dades(dades)
    pacient['simptomes'] = normalitzar_simptomes(pacient['simptomes'])
    pacient['simptomes'] = assignar_risc_simptomes(pacient['simptomes'])
    pacient['risc_total'] = calcular_risc_total(pacient['simptomes'])
    mostrar_pacient(pacient)
    pacients.append(pacient)

  pacients = ordenar_pacients(pacients)
  mostrar_llista_pacients(pacients)
```

```python
main_gestio_pacients()
```
