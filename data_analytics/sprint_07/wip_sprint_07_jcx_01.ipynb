## N1.1: Selecció de Personal

**ALGORITMO PRINCIPAL**: *main_seleccio()*

Mientras no llegamos a 10 candidaturas:
1. Obtener datos de canditatura: *obtenir_dades()*
2. Validar los datos de la candidatura: *validar_dades()*
3. Calcular la puntuación de la candidatura: *calcular_puntuacio()*
4. Mostrar el informe individual de la candidatura: *mostrar_individual()*

Al llegar a 10 candidaturas validadas:

5. Ordenar candidaturas para informe final: *crear_informe()*
6. Mostrar informe final: *mostrar_informe()*


```python
def obtenir_dades(dades_test=None):
  """"
  Esta función tiene como entrada una tupla de datos de un posible candidato.
  Si nos proporcionan datos de entrada, utilizaré esos datos.
  Si no hay datos de entrada, la función pedirá los datos por pantalla al usuario.

  Queremos incorporar los siguientes datos:
  - Nombre y Apellidos (nom)
  - Teléfono de contacto (telefono)
  - Años de experiencia (experiencia)
  - Conocimientos de Catalán (catala)
  - Dominio de Excel (excel)
  - Título Académico (grau)

  La función retornará una tupla de datos con seis campos:
  (nom, telefon, experiencia, catala, excel, grau)
  """

  if dades_test:
    print("\nCarregant les dades automàticament...")
    try:
      nom = dades_test[0]
      telefon = dades_test[1]
      experiencia = dades_test[2]
      catala = dades_test[3]
      excel = dades_test[4]
      grau = dades_test[5]
      return (nom, telefon, experiencia, catala, excel, grau)

    except IndexError:
      print("Error: Falten dades")
      return
    except:
      print("Error: Dades incorrectes")
      return

  else:
    print("\nSisplau, entra les dades manualment...")
    nom = input("Introdueix el nom: ").strip()
    telefon = input("Introdueix el telèfon: ").strip()
    experiencia = input("Introdueix el número d'anys a l'empresa: ").strip()
    catala = input("Introdueix el teu nivell de català\n(0 = sense coneixement, 1 = fins a nivell B1, 2 = nivell B2, 3 = C1 o superior): ").strip()
    excel = input("Introdueix el teu nivell d'Excel\n(0 = nivell bàsic, 1 = ús de funcions, 2 = taules dinàmiques, 3 = Visual Basic i Macros): ").strip()
    grau = input("Introdueix el teu nivell de Títol acadèmic\n(0 = ESO, 1 = FP, 2 = Universitat, 3 = Màster o superior): ").strip()
    return (nom, telefon, experiencia, catala, excel, grau)
```

```python
def validar_dades(dades_candidat):
  """
  Esta función recibe una tupla de datos
  y comprueba que estos datos sean del tipo correcto,
  para cada uno de sus campos:

  nom -> Debe ser un texto no vacío
  telefon -> Debe ser un texto no vacío
  experiencia -> Debe ser un número entero
  catala -> Debe ser un número entero entre 0 y 3
  excel -> Debe ser un número entero entre 0 y 3
  grau -> Debe ser un número entero entre 0 y 3

  Finalmente la función retorna los datos validados,
  o en caso de datos incorrectos, no retorna nada.
  """

  try:
    nom = dades_candidat[0]
    telefon = dades_candidat[1]
    experiencia = int(dades_candidat[2])
    catala = int(dades_candidat[3])
    excel = int(dades_candidat[4])
    grau = int(dades_candidat[5])
  except:
    print("Error: Dades incorrectes")
    return

  if len(nom) == 0:
    print("Error: Nom no vàlid")
    return
  if len(telefon) == 0:
    print("Error: Telèfon no vàlid")
    return

  if (catala < 0 or catala > 3) or (excel < 0 or excel > 3) or (grau < 0 or grau > 3):
    print("Error: Valors de Català / Excel / Títol Acadèmic han d'estar entre 0 i 3")
    return

  return (nom, telefon, experiencia, catala, excel, grau)
```

```python
def calcular_puntuacio(experiencia, catala, excel, grau):
  """
  Esta función recibe cuatro argumentos de entrada de tipo entero:
  - Años de experiencia: experiencia
  - Nivel de catalán: catala
  - Dominio de Excel: excel
  - Título académico: grau

  La función retorna un número según la fórmula:
  puntuación = 1.5 * experiencia + catala + excel + grau
  """

  if (experiencia < 0):
    print("Error: El número d'anys d'experiència ha de ser major que 0")
    return

  if (catala < 0 or catala > 3) or (excel < 0 or excel > 3) or (grau < 0 or grau > 3):
    print("Error: Valors de Català / Excel / Títol Acadèmic han d'estar entre 0 i 3")
    return

  return 1.5 * experiencia + catala + excel + grau
```

```python
def classificar_candidat(puntuacio):
  """
  Con el valor de entrada proporcionado del parámetro *puntuacio* (tipo float),
  la función retorna un texto según la clasificación del candidato:

  'Prioritat Alta' si puntuacio > 12
  'Candidat Adequat' si 8 ≤ puntuacio < 12
  'No Prioritari' si puntuacio < 8
  """

  if puntuacio > 12:
    return "Alta Prioritat"
  elif 8 <= puntuacio < 12:
    return "Candidat Adequat"
  else:
    return "No Prioritari"
```

```python
def mostrar_individual(dades_canditat):
  """
  Esta función recibe como entrada un diccionario
  con los datos ya validados correctamente de un posible candidato.
  El diccionario incluye las siguientes claves y valores:

  - 'nom' -> text con el nombre y apellidos
  - 'telefon' -> teléfono como string
  - 'experiencia' -> número entero con los años de experiencia
  - 'catala' -> número entero con el nivel de catalán
  - 'excel' -> número entero con el nivel de Excel
  - 'grau' -> número entero con el nivel de grado académico
  - 'puntuacio' -> número float con la puntuación del candidato
  - 'classificacio' -> string con la valoración del candidato

  Se muestra por pantalla un texto con los datos del candidato.
  """

  informe = ""

  try:
    for key, value in dades_canditat.items():
      if key == 'nom':
        informe += f"Nom i Cognoms: {value}\n"
      elif key == 'telefon':
        informe += f"Telèfon: {value}\n"
      elif key == 'experiencia':
        informe += f"Anys d'experiència a l'empresa: {value}\n"

      elif key == 'catala':
        nivell_catala = ''
        if value == 0:
          nivell_catala = 'Sense coneixement'
        elif value == 1:
          nivell_catala = 'Nivell B1'
        elif value == 2:
          nivell_catala = 'Nivell B2'
        elif value == 3:
          nivell_catala = 'C1 o superior'
        informe += f"Coneixement de català: {nivell_catala}\n"

      elif key == 'excel':
        nivell_excel = ''
        if value == 0:
          nivell_excel = 'Nivell bàsic'
        elif value == 1:
          nivell_excel = 'Ús de funcions'
        elif value == 2:
          nivell_excel = 'Taules dinàmiques'
        elif value == 3:
          nivell_excel = 'Visual Basic i Macros'
        informe += f"Domini d'Excel: {nivell_excel}\n"

      elif key == 'grau':
        nivell_academic = ''
        if value == 0:
          nivell_academic = 'ESO'
        elif value == 1:
          nivell_academic = 'FP'
        elif value == 2:
          nivell_academic = 'Universitat'
        elif value == 3:
          nivell_academic = 'Màster o superior'
        informe += f"Títol Acadèmic: {nivell_academic}\n"

      elif key == 'puntuacio':
        informe += f"Puntuació: {value}\n"
      elif key == 'classificacio':
        informe += f"Classificació: {value}"

    print(informe)

  except:
    print("Error creant infore: dades incorrectes")
```

```python
def mostrar_informe(candidatures):
  """
  Esta función recibe una lista ordenada con las 10 candidaturas seleccionadas.
  Es una lista de diccionarios que incluyen los datos de cada candidatura
  - 'nom' -> text con el nombre y apellidos
  - 'telefon' -> teléfono como string
  - 'puntuacio' -> número float con la puntuación

  Se muestra un texto con el informe formateado de los datos de los candidatos.
  """
  informe = '\nINFORME FINAL\n'
  for i in range(len(candidatures)):
    informe += f"{i+1}) Nom: {candidatures[i]['nom']}  Telèfon: {candidatures[i]['telefon']}  Puntuació: {candidatures[i]['puntuacio']}\n"

  print(informe)
```

**ALGORITMO PRINCIPAL**: *main_seleccio()*

Mientras no llegamos a 10 candidaturas:
1. Obtener datos de canditatura: *obtenir_dades()*
2. Validar los datos de la candidatura: *validar_dades()*
3. Calcular la puntuación de la candidatura: *calcular_puntuacio()*
4. Calcular clasificación de candidatura: *classificar_candidat()*
5. Añadir candidato a la lista de candidaturas
6. Mostrar el informe individual de la candidatura: *mostrar_individual()*

Al llegar a 10 candidaturas validadas:

7. Ordenar candidaturas para informe final: *crear_informe()*
8. Mostrar informe final: *mostrar_informe()*


```python
def main_seleccio():

  llista_candidats = []

  inputs_test = [
    "Laura Serra", "608112233", 6, 3, 3, 3,
    "Clara Gómez", "655221199", 3, 3, 3, 2,
    "Sergio López", "644112211", 3, 2, 2, 2,
    "Candidat Error", "611000000", 5, 7, 3, 2, # Català = 7 → NO VÀLID
    "Marta Ruiz", "699887766", 2, 3, 2, 1,
    "Marc Vidal", "612443322", 6, 3, 3, 3,
    "Júlia Romero", "677998877", 5, 3, 3, 2,
    "Pol Navarro", "699554433", 4, 3, 2, 2,
    "David Ferrer", "677665544", 2, 1, 1, 1,
    "Anna Paredes", "633221144", 1, 2, 1, 0,
    "Anibal Paredes", "644272727", 1, 2, 1, 0,
    "Núria Casals", "600112200", 8, 2, 3, 3 # 11è candidat
    ]

  while len(llista_candidats) < 10:
    # 1
    candidat = obtenir_dades(inputs_test[:6])
    del inputs_test[:6]

    if candidat:
      # 2
      candidat_validat = validar_dades(candidat)

    if candidat_validat:
      dicc_candidat = {}
      dicc_candidat['nom'] = candidat_validat[0]
      dicc_candidat['telefon'] = candidat_validat[1]
      dicc_candidat['experiencia'] = candidat_validat[2]
      dicc_candidat['catala'] = candidat_validat[3]
      dicc_candidat['excel'] = candidat_validat[4]
      dicc_candidat['grau'] = candidat_validat[5]

      # 3
      dicc_candidat['puntuacio'] = calcular_puntuacio(candidat_validat[2], candidat_validat[3], candidat_validat[4], candidat_validat[5])
      # 4
      dicc_candidat['classificacio'] = classificar_candidat(dicc_candidat['puntuacio'])
      # 5
      llista_candidats.append(dicc_candidat)
      # 6
      print("\nINFORME CANDIDATURA", len(llista_candidats))
      mostrar_individual(dicc_candidat)

      if len(llista_candidats) < 10:
        input("\nPrem ENTER per entrar següent candidat...")
      else:
        input("\nPrem ENTER per veure informe final...")

  # 7
  llista_candidats.sort(key=lambda x: (x['classificacio'], -x['puntuacio']))
  # 8
  mostrar_informe(llista_candidats)
```

```python
def obtenir_segment_i_factor(client):
  """
  Esta función recibe como entrada un diccionario con los datos iniciales de un cliente.

  Incorporaremos al cliente los campos (clave / valor) de:
  - 'segment': Un string con la combinación del tipo de comportamiento con el grupo de edad (ej. Navega pero no añade productos / 18-25)
  - 'factor_codi': Número float correspondiente según su código de comportamiento (lo obtenemos del diccionario_marketing)

  La función retorna un diccionario con los datos actualizados del cliente.
  """

  cliente_actualizado = client.copy()
  cliente_actualizado['segment'] = dicc_marketing['comportaments'][client['codi_comportament']] + " / " + client['grup_edat']
  cliente_actualizado['factor_codi'] = dicc_marketing['factors_comportament'][client['codi_comportament']]

  return cliente_actualizado
```
