**ALGORITMO PRINCIPAL**: *main_seleccio()*

Mientras no llegamos a 10 candidaturas:
1. Obtener datos de canditatura: *obtenir_dades()*
2. Validar los datos de la candidatura: *validar_dades()*
3. Calcular la puntuación de la candidatura: *calcular_puntuacio()*
4. Calcular clasificación de candidatura: *classificar_candidat()*
5. Añadir candidato a la lista de candidaturas
6. Mostrar el informe individual de la candidatura: *mostrar_individual()*
7. Entrar siguiente candidatura

Al llegar a 10 candidaturas validadas:

8. Ordenar candidaturas para informe final: *crear_informe()*
9. Mostrar informe final: *mostrar_informe()*


```python
def obtenir_dades(dades_test=None):
  """"
  Esta función recoge los datos de un posible candidato.
  Si nos proporcionan datos de entrada, utilizaré esos datos.
  Si no hay datos de entrada, la función pedirá los datos por pantalla al usuario.

  Args:
  dades_test (str, str, int, int, int, int): Tupla de datos con los campos
  (nom, telefon, experiencia, catala, excel, grau)

  Returns:
  La función retornará una tupla de datos con 6 campos:
  (nom, telefon, experiencia, catala, excel, grau)
  """

  # Cargo datos test (si existen)
  if dades_test:
    print("\nCarregant les dades automàticament...")
    try:
      nom = dades_test[0]
      telefon = dades_test[1]
      experiencia = dades_test[2]
      catala = dades_test[3]
      excel = dades_test[4]
      grau = dades_test[5]
      return (nom, telefon, experiencia, catala, excel, grau)

    except IndexError:
      print("Error: Falten dades")
      return
    except:
      print("Error: Dades incorrectes")
      return

  # Cargo datos manualmente si es necesario
  else:
    print("\nSisplau, entra les dades manualment...")
    try:
      nom = input("Introdueix el nom: ").strip()
      telefon = input("Introdueix el telèfon: ").strip()
      experiencia = int(input("Introdueix el número d'anys a l'empresa: ").strip())
      catala = int(input("Introdueix el teu nivell de català\n(0 = sense coneixement, 1 = fins a nivell B1, 2 = nivell B2, 3 = C1 o superior): ").strip())
      excel = int(input("Introdueix el teu nivell d'Excel\n(0 = nivell bàsic, 1 = ús de funcions, 2 = taules dinàmiques, 3 = Visual Basic i Macros): ").strip())
      grau = int(input("Introdueix el teu nivell de Títol acadèmic\n(0 = ESO, 1 = FP, 2 = Universitat, 3 = Màster o superior): ").strip())
      return (nom, telefon, experiencia, catala, excel, grau)
    except ValueError:
      print("Error: Sispalu entra els valors correctament")
      return
```

```python
def validar_dades(dades_candidat):
  """
  Esta función recibe una tupla de datos
  y comprueba que estos datos sean correctos.

  Args:
  dades_candidat (tuple): Una tupla de datos con seis campos:
  nom -> Debe ser un texto no vacío
  telefon -> Debe ser un texto no vacío
  experiencia -> Debe ser un número entero
  catala -> Debe ser un número entero entre 0 y 3
  excel -> Debe ser un número entero entre 0 y 3
  grau -> Debe ser un número entero entre 0 y 3

  Returns:
  Finalmente la función retorna True si los datos están correctos,
  o en caso de datos incorrectos, retorna False.
  """

  try:
    nom = dades_candidat[0]
    telefon = dades_candidat[1]
    experiencia = dades_candidat[2]
    catala = dades_candidat[3]
    excel = dades_candidat[4]
    grau = dades_candidat[5]
  except:
    print("Error: Dades incorrectes")
    return False

  if len(nom) == 0:
    print("Error: Nom no vàlid")
    return False
  if len(telefon) == 0:
    print("Error: Telèfon no vàlid")
    return False

  if (catala < 0 or catala > 3) or (excel < 0 or excel > 3) or (grau < 0 or grau > 3):
    print("Error: Valors de Català / Excel / Títol Acadèmic han d'estar entre 0 i 3")
    return False

  return True
```

```python
def calcular_puntuacio(experiencia, catala, excel, grau):
  """
  Calcula la puntuación de un posible candidato.

  Args:
  Esta función recibe cuatro argumentos de entrada de tipo entero:
  - Años de experiencia: experiencia (int)
  - Nivel de catalán: catala (int)
  - Dominio de Excel: excel (int)
  - Título académico: grau (int)

  Returns: puntuacio (float)
  La función retorna un número según la fórmula:
  puntuación = 1.5 * experiencia + catala + excel + grau
  """

  if (experiencia < 0):
    print("Error: El número d'anys d'experiència ha de ser major que 0")
    return

  if (catala < 0 or catala > 3) or (excel < 0 or excel > 3) or (grau < 0 or grau > 3):
    print("Error: Valors de Català / Excel / Títol Acadèmic han d'estar entre 0 i 3")
    return

  return 1.5 * experiencia + catala + excel + grau
```

```python
def classificar_candidat(puntuacio):
  """
  Se clasifica un posible candidato según su puntuación.

  Args: puntuacio (float)
  Valor de entrada proporcionado del parámetro puntuacio (tipo float).

  Returns: (str)
  La función retorna un texto según la clasificación del candidato:
  'Alta Prioritat' si puntuacio > 12
  'Candidat Adequat' si 8 ≤ puntuacio < 12
  'No Prioritari' si puntuacio < 8
  """

  if puntuacio > 12:
    return "Alta Prioritat"
  elif 8 <= puntuacio < 12:
    return "Candidat Adequat"
  else:
    return "No Prioritari"
```

```python
def mostrar_individual(dades_canditat):
  """
  Se muestra por pantalla un texto con los datos del candidato.

  Args: dades_canditat (dict)
  La función recibe como entrada un diccionario
  con los datos ya validados correctamente de un posible candidato.
  El diccionario incluye las siguientes claves y valores:

  - 'nom' -> text con el nombre y apellidos
  - 'telefon' -> teléfono como string
  - 'experiencia' -> número entero con los años de experiencia
  - 'catala' -> número entero con el nivel de catalán
  - 'excel' -> número entero con el nivel de Excel
  - 'grau' -> número entero con el nivel de grado académico
  - 'puntuacio' -> número float con la puntuación del candidato
  - 'classificacio' -> string con la valoración del candidato
  """

  informe = ""

  try:
    for key, value in dades_canditat.items():
      if key == 'nom':
        informe += f"Nom i Cognoms: {value}\n"
      elif key == 'telefon':
        informe += f"Telèfon: {value}\n"
      elif key == 'experiencia':
        informe += f"Anys d'experiència a l'empresa: {value}\n"

      elif key == 'catala':
        nivell_catala = ''
        match value:
          case 0:
            nivell_catala = 'Sense coneixement'
          case 1:
              nivell_catala = 'Nivell B1'
          case 2:
              nivell_catala = 'Nivell B2'
          case 3:
              nivell_catala = 'C1 o superior'
          case _:
              nivell_catala = 'Nivell desconegut'
        informe += f"Coneixement de català: {nivell_catala}\n"

      elif key == 'excel':
        nivell_excel = ''
        match value:
          case 0:
            nivell_excel = 'Nivell bàsic'
          case 1:
            nivell_excel = 'Ús de funcions'
          case 2:
            nivell_excel = 'Taules dinàmiques'
          case 3:
            nivell_excel = 'Visual Basic i Macros'
          case _:
            nivell_excel = 'Nivell desconegut'
        informe += f"Domini d'Excel: {nivell_excel}\n"

      elif key == 'grau':
        nivell_academic = ''
        match value:
          case 0:
            nivell_academic = 'ESO'
          case 1:
            nivell_academic = 'FP'
          case 2:
            nivell_academic = 'Universitat'
          case 3:
            nivell_academic = 'Màster o superior'
          case _:
            nivell_academic = 'Nivell desconegut'
        informe += f"Títol Acadèmic: {nivell_academic}\n"

      elif key == 'puntuacio':
        informe += f"Puntuació: {value}\n"
      elif key == 'classificacio':
        informe += f"Classificació: {value}"

    print(informe)

  except:
    print("Error creant infore: dades incorrectes")
```

```python
def mostrar_informe(candidatures):
  """
  Se muestra un texto con el informe de los datos de 10 candidaturas.

  Args: candidatures (list)

  Esta función recibe una lista ordenada con las 10 candidaturas seleccionadas.
  Es una lista de diccionarios que incluyen los datos de cada candidatura:
  - 'nom' -> texo con el nombre y apellidos
  - 'telefon' -> teléfono como string
  - 'puntuacio' -> número float con la puntuación del candidato
  - 'classificacio' -> string con la valoración del candidato
  """
  informe = '\nINFORME FINAL\n'
  for i in range(len(candidatures)):
    informe += f"{i+1:<2}) {candidatures[i]['nom']:>15}  Telèfon: {candidatures[i]['telefon']:12}  Puntuació: {candidatures[i]['puntuacio']}\n"

  print(informe)
```

```python
def main_seleccio():

  # Datos test
  inputs_test = [
    "Laura Serra", "608112233", 6, 3, 3, 3,
    "Clara Gómez", "655221199", 3, 3, 3, 2,
    "Sergio López", "644112211", 3, 2, 2, 2,
    "Candidat Error", "611000000", 5, 7, 3, 2, # Català = 7 → NO VÀLID
    "Marta Ruiz", "699887766", 2, 3, 2, 1,
    "Marc Vidal", "612443322", 6, 3, 3, 3,
    "Júlia Romero", "677998877", 5, 3, 3, 2,
    "Pol Navarro", "699554433", 4, 3, 2, 2,
    "David Ferrer", "677665544", 2, 1, 1, 1,
    "Anna Paredes", "633221144", 1, 2, 1, 0,
    "Anibal Paredes", "644272727", 1, 2, 1, 0,
    "Núria Casals", "600112200", 8, 2, 3, 3 # 11è candidat
    ]

  llista_candidats = []

  while len(llista_candidats) < 10:
    # 1. Obtener datos de canditatura
    candidat = obtenir_dades(inputs_test[:6])
    del inputs_test[:6]

    # 2. Validar los datos de la candidatura
    if validar_dades(candidat):
      dicc_candidat = {}
      dicc_candidat['nom'] = candidat[0]
      dicc_candidat['telefon'] = candidat[1]
      dicc_candidat['experiencia'] = candidat[2]
      dicc_candidat['catala'] = candidat[3]
      dicc_candidat['excel'] = candidat[4]
      dicc_candidat['grau'] = candidat[5]

      # 3. Calcular la puntuación de la candidatura
      dicc_candidat['puntuacio'] = calcular_puntuacio(candidat[2], candidat[3], candidat[4], candidat[5])
      # 4. Calcular clasificación de candidatura
      dicc_candidat['classificacio'] = classificar_candidat(dicc_candidat['puntuacio'])

      # 5. Añadir candidato a la lista de candidaturas
      llista_candidats.append(dicc_candidat)

      # 6. Mostrar el informe individual de la candidatura
      print("\nINFORME CANDIDATURA", len(llista_candidats))
      mostrar_individual(dicc_candidat)

    # 7. Entrar siguiente candidatura
    if len(llista_candidats) < 10:
      input("\nPrem ENTER per entrar següent candidat...")
    else:
      input("\nPrem ENTER per veure informe final...")

  # 8. Ordenar candidaturas para informe fina
  llista_candidats.sort(key=lambda x: (x['classificacio'], -x['puntuacio']))

  # 9. Mostrar informe final
  mostrar_informe(llista_candidats)
```

```python
# Construyo este diccionario de Marketing (diccionario de diccionarios)
# con los códigos y valores que nos propone el enunciado,
# porque luego lo necesitaremos consultar para poder procesar cada cliente
def diccionari_marketing():
  dicc_marketing = { 'comportaments': { 0: "Navega pero no añade productos",
                                        1: "Añade productos pero no llega al checkout",
                                        2: "Llega al checkout pero no paga",
                                        3: "Cesta de la compra abandonada hace menos de 24h",
                                        4: "Cesta de la compra abandonada hace más de 24h"},

                    'factors_comportament': {  0: 0.10,
                                                1: 0.25,
                                                2: 0.50,
                                                3: 0.75,
                                                4: 0.90 },

                    'factors_edat': {  '18-25': 1.20,
                                        '26-40': 1.00,
                                        '41-60': 0.85,
                                        '60+': 0.70 }}

  return dicc_marketing
```

```python
def obtenir_segment_i_factor(client, dicc_marketing):
  """
  Se actualizan los datos de un cliente añadiendo los campos de:
  - 'segment': Un string con la combinación del tipo de comportamiento con el grupo de edad (ej. Navega pero no añade productos / 18-25)
  - 'factor_codi': Número float correspondiente según su código de comportamiento (lo obtenemos del diccionario_marketing)

  Args: client (dict), dicc_marketing (dict)
  La función recibe como entrada un diccionario con los datos de un cliente.
  Y también el diccionario de marketing.

  Returns: cliente_actualizado (dict)
  La función retorna un diccionario con los datos actualizados del cliente.
  """

  cliente_actualizado = client.copy()
  cliente_actualizado['segment'] = dicc_marketing['comportaments'][client['codi_comportament']] + " / " + client['grup_edat']
  cliente_actualizado['factor_codi'] = dicc_marketing['factors_comportament'][client['codi_comportament']]

  return cliente_actualizado
```

```python
def obtenir_factor_edat(client, dicc_marketing):
  """
  Se incorpora a los datos de un cliente su factor de edad según el diccionario de Marketing.

  Args: client (dict), dicc_marketing (dict)
  Returns: factor_edat (float)
  """
  return dicc_marketing['factors_edat'][client['grup_edat']]
```

```python
def calcular_risc(factor_comportament, factor_grup_edat):
  """
  Se calcula para un cliente su factor de riesgo según la fórmula:
  riesgo = factor_comportament * factor_grup_edat

  Args: factor_comportament (float), factor_grup_edat (float)
  Returns: riesgo (float)
  """
  return round(factor_comportament * factor_grup_edat, 2)
```

```python
def redactar_missatge(risc, nom):
  """
  Se redacta un mensaje personalizado según el riesgo del cliente.

  Args: risc (float), nom (str)
  Esta función recibe un float con el factor de riesgo y el nombre del cliente.

  Returns: missatge (str)
  Retorna un string con el mensaje de texto adecuado para su segmento.

  Haciendo una primera inspección de los valores posibles de riesgo y segmento,
  propongo estos posibles casos distintos para personalizar los mensajes:

  - No añaden productos -> (riesgo <= 0.15) =- missatge_0
  - No llegan al checkout -> (0.15 < riesgo <= 0.30) -> missatge_1
  - No llegan a pagar -> (0.30 < riesgo <= 0.60) -> missatge_2
  - Cesta de la compra abandonada -> (0.60 < riesgo <= 0.80) -> missatge_3
  - Cesta de la compra abandonada (jóvenes) -> (riesgo > 0.80) -> missatge_4
  """

  missatge_0 = "Gracias por su visita:\nEsperemos encuentre todo lo que necesita en nuestra tienda."
  missatge_1 = "Gracias por su visita:\nRecuerde hacer checkout para procesar su compra."
  missatge_2 = "Gracias por comprar en nuestra tienda:\nRecuerde completar el checkout para poder procesar su pedido."
  missatge_3 = "Gracias por comprar en nuestra tienda:\nPor favor recuerde hacer checkout para poder completar su compra."
  missatge_4 = f"Buenos días {nom}:\nEstamos muy contentos de contar contigo como cliente.\nIMPORTANTE: por favor, recuerda hacer el checkout para poder completar tu compra.\nMuchas gracias, deseamos que encuentres todo lo que deseas en nuestra tienda!"

  match risc:
    case x if x <= 0.15:
      return missatge_0
    case x if 0.15 < x <= 0.30:
      return missatge_1
    case x if 0.30 < x <= 0.60:
      return missatge_2
    case x if 0.60 < x <= 0.80:
      return missatge_3
    case x if x > 0.80:
      return missatge_4
    case _:
      return missatge_0
```

```python
def processar_clients(llista_clients, dicc_marketing):
  """
  Se procesan los datos de los clientes, añadiendo diferentes campos.

  Args: llista_clients (list), dicc_marketing (dict)
  Esta función recibe como entrada una lista de diccionarios con los datos iniciales de cada cliente.
  También recibe el diccionario de marketing.

  Cada diccionario viene con los siguientes campos de clave / valor:
    - 'id_client': Número entero con el identificador único de cada cliente
    - 'nom': String con el nombre
    - 'edat': Número entero con la edad
    - 'grup_edat': String con el grupo de edad
    - 'codi_comportament': Número entero con el código de comportamiento según los valores:
        0 -> Navega pero no añade productos
        1 -> Añade productos pero no llega al checkout
        2 -> Llega al checkout pero no paga
        3 -> Cesta de la compra abandonada hace menos de 24h
        4 -> Cesta de la compra abandonada hace más de 24h

  Al procesar cada cliente, debemos incorporar los siguientes campos:
    - 'segment': Un string con la combinación del tipo de comportamiento con el grupo de edad (ej. Navega pero no añade productos / 18-25)
    - 'factor_codi': Número float correspondiente según su código de comportamiento (lo obtenemos del diccionario_marketing)
    - 'factor_edat': Número float correspondiente según su grupo de edad (lo obtenemos del diccionario_marketing)
    - 'risc_final': Número float con el cálculo del riesgo según la fórmula risc_final = factor codi * factor edat
    - 'missatge': Un string con el mensaje de texto adecuado para su segmento

  Returns: clients_mktg (list)
  Se retorna una lista con los datos actualizados de los clientes.
  """
  clients_mktg = []

  for client in llista_clients:
    client_plus = client.copy()
    client_plus = obtenir_segment_i_factor(client, dicc_marketing)
    client_plus['factor_edat'] = obtenir_factor_edat(client, dicc_marketing)
    client_plus['risc_final'] = calcular_risc(client_plus['factor_codi'], client_plus['factor_edat'])
    client_plus['missatge'] = redactar_missatge(client_plus['risc_final'], client_plus['nom'])
    clients_mktg.append(client_plus)

  return clients_mktg
```

Finalmente:

```python
def obtenir_dades_pacients():
  """
  Recogemos los datos de los clientes que se nos proporcionan.

  Returns: pacients (list)
  Retornamos una lista de strings con los datos de los pacientes.
  """

  pacients = [
      "Maria|45|tos, Fiebre Alta, dolor_pecho",
      "Luis|33|dolor_cabeza,faTiga",
      "Sara|67|fiebre alta, dificultad RESPIRAR, tos",
      "Jordi|52|fatiga , tos",
      "Anna|29|fiebre alta,dolor_cabeza"
  ]

  return pacients
```

```python
def convertir_dades(dades):
  """
  Se validan y formatean los datos de un paciente.

  Args: dades (str)
  Esta función tiene como entrada un string con los datos en bruto de un paciente (separados mediante |).

  Returns: pacient (dict)
  Retorna un diccionario con los datos del paciente procesados.
  """
  try:
    nom, edat, simptomes = dades.split("|")

    return {
        'nom': nom,
        'edat': edat,
        'simptomes': simptomes
    }

  except ValueError:
    print("Error: Datos incorrectos del paciente.")
    return None

  except AttributeError:
    print("Error: Datos de entrada incorrectos.")
    return None
```

```python
convertir_dades('Maria|45|tos, Fiebre Alta, dolor_pecho')
```

```python
def normalitzar_simptomes(simptomes):
  """
  Se convierte el campo de simptomes de un paciente a una lista formateada de forma correcta.
  - Quitamos espacios
  - Convertimos a minúsculas
  - Convertimos los espacios en guiones bajos

  Args: simptomes (str)
  Returns: simptomes_ok (list)
  """

  simptomes = simptomes.strip().lower()

  simptomes_ok = []
  for simptoma in simptomes.split(","):
    simptoma = simptoma.strip().replace(" ", "_")
    simptomes_ok.append(simptoma)

  return simptomes_ok
```

```python
normalitzar_simptomes('tos, Fiebre Alta, dolor_pecho')
```

```python
def assignar_risc_simptomes(simptomes):
  """
  Convertimos la lista de simptomes en una lista de tuplas con el síntoma y su valor de riesgo.

  Args: simptomes (list)
  Se recibe como entrada una lista de simptomes ya normalizada.
  Cada elemento de la lista de simptomes es un string con el nombre correcto.
  Convertiremos cada elemento de la lista en una tupla con dos elementos:
  (simptoma, valor de riesgo según el diccionario nivells_risc)

  Returns: simptomes_risc (list)
  Retornamos la lista de tuplas.
  """

  nivells_risc = {
      "fiebre_alta": 3,
      "dificultad_respirar": 5,
      "dolor_pecho": 4,
      "tos": 1,
      "fatiga": 1,
      "dolor_cabeza": 1
  }

  simptomes_risc = []

  for simptoma in simptomes:
    if simptoma in nivells_risc:
      simptomes_risc.append((simptoma, nivells_risc[simptoma]))
    else:
      simptomes_risc.append((simptoma, 0))

  return simptomes_risc
```

```python
assignar_risc_simptomes(['tos', 'fiebre_alta', 'dolor_pecho'])
```

```python
def calcular_risc_total(simptomes):
  """
  Se calcula el riesgo correspondiente según los síntomas de un paciente.

  Args: simptomes (list)
  Esta función recibe una lista de tuplas de simptomas y sus respectivos valores de riesgo.

  Returns: risc_total (float)
  Retorna el valor de riesgo total, sumando los valores de cada síntoma.
  """
  risc_total = 0

  for simptoma_risc in simptomes:
    if simptoma_risc[1] > 0:
      risc_total += simptoma_risc[1]

  return risc_total
```

```python
calcular_risc_total([('tos', 1), ('fiebre_alta', 3), ('dolor_pecho', 4)])
```

```python
def mostrar_pacient(pacient):
  """
  Se muestra por pantalla los datos de un paciente.
  Args: pacient (dict)
  Se recibe como entrada un diccionario con los datos de un paciente.
  """
  print(f"Nombre: {pacient['nom']}")
  print(f"Edad: {pacient['edat']}")

  print("Síntomas:")
  simptomes_pacient = ""
  for simptoma in pacient['simptomes']:
    simptomes_pacient += f"{simptoma[0]} ({simptoma[1]}), "
  print(simptomes_pacient[:-2])

  print(f"Riesgo total: {pacient['risc_total']}")
  print()
```

```python
def ordenar_pacients(pacients):
  """
  Se ordenan los pacientes según su riesgo de mayor a menor.

  Args: pacients (list)
  Una lista de diccionarios con los datos de los pacientes.

  Returns: pacients (list)
  Lista de pacientes ya ordenada.
  """
  pacients.sort(key=lambda x: x['risc_total'], reverse=True)
  return pacients
```

```python
def main_gestio_pacients():
  # 1. Obtener datos de pacientes
  llista_dades = obtenir_dades_pacients()
  pacients = []

  for dades in llista_dades:
    # 2. Convertir datos de paciente
    pacient = convertir_dades(dades)
    # 3. Limpiar datos de síntomas
    pacient['simptomes'] = normalitzar_simptomes(pacient['simptomes'])
    # 4. Asignar valores de riesgo a cada síntoma
    pacient['simptomes'] = assignar_risc_simptomes(pacient['simptomes'])
    # 5. Calcular valor de riesgo total
    pacient['risc_total'] = calcular_risc_total(pacient['simptomes'])
    # 6. Mostrar datos correctos del paciente
    mostrar_pacient(pacient)
    # 7. Añadir a la lista de pacientes validados
    pacients.append(pacient)

  # 8. Ordenar la lista de pacientes por prioridad
  pacients = ordenar_pacients(pacients)

  # 9. Mostrar lista ordenada de pacientes
  print("LISTA DE PACIENTES SEGÚN PRIORIDAD:")
  mostrar_llista_pacients(pacients)
```

```python
def decidir_longitud(default=12):
  """
  Pregunta al usuario la longitud deseado de la contraseña.
  Si no contesta nada, coge el valor por defecto.

  Args:
  default (int): Longitud por defecto de la contraseña (12).

  Returns:
  longitud (int): Longitud de la contraseña.
  """

  while True:
    entrada = input(f'Cuántos caracteres quieres? (entre 4 y 99): ')
    if entrada == "":
      return default

    try:
      longitud = int(entrada)
      if 3 < longitud < 100:
        return longitud
      else:
        print("Error: Tienes que introducir un número entre 4 y 99\n")
    except ValueError:
        print("Error: Tienes que introducir un número válido\n")
```

```python
def decidir_segons_input(tema):
  """
  Pregunta al usuario su decisión afirmativa o negativa sobre un tema.

  Args:
  tema (str): El tema específico que se insertará en la pregunta.

  Returns:
  - True -> si la respuesta es afirmativa ('s', 'sí'...)
  - False -> si es negativa ('n', 'no'...)
  - None -> si el input es ambiguo
  """

  pregunta = f"Quieres que la contraseña incluya {tema}? (s/n): "
  decisio = input(pregunta).strip().lower()

  if decisio in ('s', 'si', 'sí', 'yes', 'y', 'oui', 'ok'):
    return True
  elif decisio in ('n', 'no', 'nah', 'never'):
    return False
  else:
    print("Tienes que introducir s o n. Usando valor por defecto\n")
    return None
```

```python
def generar_minuscula():
  """
  Genera una letra minúscula al azar.
  Returns: un string de un solo carácter con la letra minúscula.
  """
  return np.random.choice(list(string.ascii_lowercase))
```

```python
def barrejar_contrassenya(contrassenya):
  """
  Reordena aleatoriamente los caracteres de una contraseña.

  Args:
  contrassenya (str): La contraseña a reordenar.

  Returns:
  Contraseña reordenada (str).
  """
  llista_caracters = list(contrassenya)
  np.random.shuffle(llista_caracters)
  return "".join(llista_caracters)
```

```python
def crear_contrassenya(longitud, majuscules=True, minuscules=True, numeros=True, simbols=False):
  """
  Crea una contraseña aleatoria con los parámetros indicados.

  Por decisión programática, establecemos que tenga una longitud mínima de 4 y máxima de 99.
  También nos aseguramos que al menos haya algún tipo de carácter (que no todo sea False)
  y que haya al menos un carácter de cada tipo de los que hemos escogido.

  Args:
  longitud (int) -> Longitud de la contraseña.
  majuscules (bool) -> Si se incluyen mayúsculas.
  minuscules (bool) -> Si se incluyen minúsculas.
  numeros (bool) -> Si se incluyen números.
  simbols (bool) -> Si se incluyen símbolos.

  Returns:
  contrassenya (str): La contraseña creada.
  """

  print(f"\nCreando contraseña con caracteres ({longitud}), mayúsculas ({majuscules}), minúsculas ({minuscules}), números ({numeros}) y símbolos ({simbols})...")

  if not (majuscules or minuscules or numeros or simbols):
        print("Error: Tiene que haber al menos algún tipo de carácter.")
        return None

  if longitud < 4:
    longitud = 4
    print("Ajustada longitud de contraseña a 4 caracteres")
  elif longitud > 99:
    longitud = 99
    print("Ajustada longitud de contraseña a 99 caracteres")

  # Generamos al menos un carácter de los tipos escogidos
  contrassenya = ""
  contrassenya += generar_majuscula() if majuscules else ""
  contrassenya += generar_minuscula() if minuscules else ""
  contrassenya += generar_numero() if numeros else ""
  contrassenya += generar_simbol() if simbols else ""

  # Creamos una lista con las funciones que usaremos para generar caracteres aleatorios
  funcions_random = []
  if majuscules:
    funcions_random.append(generar_majuscula)
  if minuscules:
    funcions_random.append(generar_minuscula)
  if numeros:
    funcions_random.append(generar_numero)
  if simbols:
    funcions_random.append(generar_simbol)

  # Generamos los caracteres que nos faltan, aleatoriamente
  for _ in range(longitud - len(contrassenya)):
    generar_caracter = np.random.choice(funcions_random)
    contrassenya += generar_caracter()

  # Reordenamos aleatoriamente todos los caracteres de la constraseña generada
  contrassenya = barrejar_contrassenya(contrassenya)

  return contrassenya
```

## N3.2: Procesamiento Automático de Datos Deportivos

**ALGORITME PRINCIPAL**: *main_resultats_esportius()*
1. Obtener datos des del fichero: *obtenir_dades_esportives()*
2. Crear diccionarios de partidos: *crear_diccionari_partits()*
3. Calcular puntuación de partidos: *repartir_punts()*
4. Crear diccionarios de equipos: *crear_diccionari_equips()*
5. Contabilizar goles: *comptar_gols()*
6. Contabilizar puntos: *comptar_punts()*
7. Contabilizar victorias: *comptar_victories()*
8. Ordenar clasificación: *ordenar_classificacio()*
9. Mostrar resultados: *mostrar_resultats()*

```python
def obtenir_dades_esportives():
  """
  Leemos los datos deportivos des de un fichero dado.

  El fichero tiene todas las linias con la estructura:
  equipo + tabulador + resultado + tabulador + equipo

  Returns:
  Devolvemos una lista de listas con los datos de los partidos.
  Cada elemento de la lista tendrá 3 strings con esta estructura:
  [ equipo_local, resultado, equipo_visitante ]
  """

  with open("historic partits.txt", "r") as fitxer:
    llista_partits = [linia.split('\t') for linia in fitxer.read().splitlines() if linia]
    return llista_partits
```

```python
llista_partits = obtenir_dades_esportives()
print(llista_partits[:5])
```

```python
def crear_diccionari_partits(llista_partits):
  """
  Creamos un diccionario con los datos de los partidos.

  Args:
  llista_partits -> Una lista de listas con los datos de los partidos.
  Cada elemento de la lista tendrá 3 strings con esta estructura:
  [ equipo_local, resultado, equipo_visitante ]

  Returns:
  Devolvemos una lista de diccionarios con los datos de los partidos.
  """

  partits = []

  for llista_partit in llista_partits:
    if len(llista_partit) == 3:
      dicc_partit = {}
      dicc_partit['local'] = llista_partit[0]
      dicc_partit['visitant'] = llista_partit[2]

      llista_resultat = llista_partit[1].split('-')
      if len(llista_resultat) == 2:
        dicc_partit['resultat'] = llista_partit[1]
        try:
          dicc_partit['gols_local'] = int(llista_resultat[0])
          dicc_partit['gols_visitant'] = int(llista_resultat[1])
          partits.append(dicc_partit)

        except ValueError:
          print(f"Error: dades errònies pel partit {llista_partit}")
      else:
        print(f"Error: dades errònies pel partit {llista_partit}")
    else:
      print(f"Error: dades errònees pel partit {llista_partit}")

  return partits
```

```python
llista_partits = [['Manlleu', '0-1', 'Granollers'], ['Vilafranca', '4-5', 'Terrassa'], ['Olot', '5-0', 'Granollers']]
partits = crear_diccionari_partits(llista_partits)
print(partits)
```

```python
def repartir_punts(partits):
  """
  Modifica IN SITU la lista de partidos añadiendo la puntuación de cada equipo
  a los diccionarios existentes. No retorna nada.

  Args:
  partits -> Una lista de diccionarios con los datos de los partidos.
  """

  for partit in partits:

    if partit['gols_local'] > partit['gols_visitant']:
      partit['punts_local'] = 3
      partit['punts_visitant'] = 0
    elif partit['gols_local'] < partit['gols_visitant']:
      partit['punts_local'] = 0
      partit['punts_visitant'] = 3
    else:
      partit['punts_local'] = 1
      partit['punts_visitant'] = 1
```

```python
def crear_diccionari_equips(partits):
  """
  Creamos un diccionario con los datos de los equipos.

  Args:
  partits -> Una lista de diccionarios con los datos de los partidos.

  Returns:
  Devolvemos un diccionario con los datos de los equipos, con las claves:
  'nom', 'gols_fets', 'gols_rebuts', 'punts', 'victories', 'derrotes' y 'empats'
  """

  equips = []
  for partit in partits:
    if partit['local'] not in equips:
      equips.append(partit['local'])
    if partit['visitant'] not in equips:
      equips.append(partit['visitant'])

  diccionari_equips = {}
  for equip in equips:
    equip_dicc = {}
    equip_dicc['nom'] = equip
    equip_dicc['gols_fets'] = 0
    equip_dicc['gols_rebuts'] = 0
    equip_dicc['punts'] = 0
    equip_dicc['victories'] = 0
    equip_dicc['derrotes'] = 0
    equip_dicc['empats'] = 0
    diccionari_equips[equip] = equip_dicc

  return diccionari_equips
```

```python
partits = [{'local': 'Manlleu', 'visitant': 'Granollers', 'resultat': '0-1', 'gols_local': 0, 'gols_visitant': 1, 'punts_local': 0, 'punts_visitant': 3}]
equips = crear_diccionari_equips(partits)
print(equips)
```

```python
def comptar_gols(partits, equips):
  """
  Contabilizamos los goles de cada equipo a partir de los diccionarios de partidos.

  Args:
  partits -> Una lista de diccionarios con los datos de los partidos.
  equips -> Un diccionario con los datos de los equipos.

  Se actualiza in situ el diccionario de equipos, añadiendo las claves y valores:
  'gols_fets' y 'gols_rebuts'
  """

  for partit in partits:
    equips[partit['local']]['gols_fets'] += partit['gols_local']
    equips[partit['local']]['gols_rebuts'] += partit['gols_visitant']
    equips[partit['visitant']]['gols_fets'] += partit['gols_visitant']
    equips[partit['visitant']]['gols_rebuts'] += partit['gols_local']
```

```python
def comptar_punts(partits, equips):
  """
  Contabilizamos los puntos para cada equipo según los diccionarios de partidos.

  Args:
  partits -> Una lista de diccionarios con los datos de los partidos.
  equips -> Un diccionario con los datos de los equipos.

  Se actualiza in situ el diccionario de equips, añadiendo las claves y valores:
  'punts_local' y 'punts_visitant'
  """

  for partit in partits:
    equips[partit['local']]['punts'] += partit['punts_local']
    equips[partit['visitant']]['punts'] += partit['punts_visitant']
```

```python
def comptar_victories(partits, equips):
  """
  Contabilizamos, para cada equipo, las victorias, derrotas y empates
  según los datos de los diccionarios de partidos.

  Args:
  partits -> Una lista de diccionarios con los datos de los partidos.
  equips -> Un diccionario con los datos de los equipos.

  Se actualiza in situ el diccionario de equips, añadiendo las claves y valores:
  'victories', 'derrotes', 'empats'
  """

  for partit in partits:
    if partit['punts_local'] == 3:
      equips[partit['local']]['victories'] += 1
      equips[partit['visitant']]['derrotes'] += 1
    elif partit['punts_visitant'] == 3:
      equips[partit['local']]['derrotes'] += 1
      equips[partit['visitant']]['victories'] += 1
    else:
      equips[partit['local']]['empats'] += 1
      equips[partit['visitant']]['empats'] += 1
```

```python
partits = [{'local': 'Manlleu', 'visitant': 'Granollers', 'resultat': '0-1', 'gols_local': 0, 'gols_visitant': 1, 'punts_local': 0, 'punts_visitant': 3}]
equips = {'Manlleu': {'nom': 'Manlleu', 'gols_fets': 0, 'gols_rebuts': 0, 'punts': 0, 'victories': 0, 'derrotes': 0, 'empats': 0}, 'Granollers': {'nom': 'Granollers', 'gols_fets': 0, 'gols_rebuts': 0, 'punts': 0, 'victories': 0, 'derrotes': 0, 'empats': 0}}

comptar_gols(partits, equips)
comptar_punts(partits, equips)
comptar_victories(partits, equips)
print(equips)
```

```python
def ordenar_classificacio(equips):
  """
  Ordenamos la lista de equipos de mayor a menor puntuación.

  Args:
  equips -> Diccionario con los datos de los equipos.

  Returns:
  classificacio -> Lista de diccionarios de los equipos, ordenados según su puntuación.
  """

  classificacio = list(equips.values())
  classificacio.sort(key=lambda x: (x['punts'], x['gols_fets'] - x['gols_rebuts']), reverse=True)
  return classificacio
```

```python
equips = {'Manlleu': {'nom': 'Manlleu', 'gols_fets': 0, 'gols_rebuts': 1, 'punts': 0, 'victories': 0, 'derrotes': 1, 'empats': 0}, 'Granollers': {'nom': 'Granollers', 'gols_fets': 1, 'gols_rebuts': 0, 'punts': 3, 'victories': 1, 'derrotes': 0, 'empats': 0}}
classificacio = ordenar_classificacio(equips)
print(classificacio)
```

```python
def mostrar_resultats(classificacio):
  """
  Mostramos por pantalla los datos que nos piden:
  - Equipo más goleador
  - Equipo más goleado
  - Clasificación

  Args:
  classificacio -> Lista de diccionarios de los equipos, ordenados según su puntuación.

  """

  golejador = max(classificacio, key=lambda x: x['gols_fets'])
  print(f"Equip més golejador: {golejador['nom']}")

  golejat = max(classificacio, key=lambda x: x['gols_rebuts'])
  print(f"Equip més golejat: {golejat['nom']}")

  print()
  print("Classificació:")
  print()
  print(f"{'EQUIP':<20}{'PUNTS':>15}{'VICTORIES':>15}{'DERROTES':>10}{'EMPATS':>10}{'GOLS':>20}")
  print("-" * 90)
  for equip in classificacio:
    gols = f"{equip['gols_fets']}-{equip['gols_rebuts']}"
    print(f"{equip['nom']:<20}{equip['punts']:>15}{equip['victories']:>15}{equip['derrotes']:>10}{equip['empats']:>10}{gols:>20}")
```

```python
def main_resultats_esportius():

  # 1. Obtener datos des del fichero
  llista_partits = obtenir_dades_esportives()

  # 2. Crear diccionarios de partidos
  partits = crear_diccionari_partits(llista_partits)

  # 3. Calcular puntuación de partidos
  repartir_punts(partits)

  # 4. Crear diccionarios de equipos
  equips = crear_diccionari_equips(partits)

  # 5. Contabilizar goles
  comptar_gols(partits, equips)

  # 6. Contabilizar puntos
  comptar_punts(partits, equips)

  # 7. Contabilizar victorias
  comptar_victories(partits, equips)

  # 8. Ordenar clasificación
  classificacio = ordenar_classificacio(equips)

  # 9. Mostrar resultados
  mostrar_resultats(classificacio)
```
